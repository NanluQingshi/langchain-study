```typescript
// 引入环境变量
import { load } from 'dotenv';

const env = await load();

const process = {
  env,
};
```

```typescript
/*
 * @Author: NanluQingshi
 * @Date: 2026-02-04 15:11:49
 * @LastEditors: NanluQingshi
 * @LastEditTime: 2026-02-04 16:17:38
 * @Description: Runnable - invoke
 */
import { Ollama } from '@langchain/community/llms/ollama';
import { HumanMessage } from '@langchain/core/messages';
import { StringOutputParser } from '@langchain/core/output_parsers';

// 初始化本地模型
const llm = new Ollama({ 
  baseUrl: 'http://localhost:11434',
  model: 'llama3.1:8b',
});

const outputParse = new StringOutputParser();

const simpleChain = llm.pipe(outputParse);

// 运行链
await simpleChain.invoke([
  new HumanMessage('Tell me a joke')
]);
```

```typescript
// Runnable - batch
await simpleChain.batch([
  [ new HumanMessage("Tell me a joke") ],
  [ new HumanMessage("Hi, Who are you?") ],
]);
```

```typescript
/*
 * @Author: NanluQingshi
 * @Date: 2026-02-04 15:11:49
 * @LastEditors: NanluQingshi
 * @LastEditTime: 2026-02-04 20:11:17
 * @Description: Runnable - fallback
 */
import { Ollama } from '@langchain/community/llms/ollama';

const fakeLLM = new Ollama({
  baseUrl: 'http://localhost:11434',
  model: 'llama3.1',
});

// await fakeLLM.invoke('你好');

const realLLM = new Ollama({
  model: 'llama3.1:8b' 
});
const llmWithFallback = fakeLLM.withFallBacks({
  fallbacks: [realLLM],
});
await llmWithFallback.invoke('你好');
```
